```python
from tabulate import tabulate
import pandas as pd

# Membaca data
df = pd.read_csv('source-marketing_data/ElectronicsProductsPricingData.csv')

# Memilih satu data saja (contoh baris pertama)
df_single = df.loc[[0]]  # Pilih baris dengan indeks 0
# Alternatif: df.iloc[[0]]

# Transpose dataframe
df_transposed = df_single.transpose()

# Menampilkan satu data sebagai tabel
print(tabulate(df_transposed, headers=["Field", "Value"], tablefmt="grid"))
```
